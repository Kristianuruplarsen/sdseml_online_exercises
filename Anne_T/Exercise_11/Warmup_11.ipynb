```python
from sympy.abc import x
```

```python
def TnFactory(n):
    if n == 0:
        return 1
    if n == 1:
        return x
    #Recursive case
    else:
        return 2 * x * TnFactory(n-1) - TnFactory(n-2)
   
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,5)
plt.scatter(x,TnFactory(5))
plt.xlabel('x', fontsize=12)
plt.ylabel('Chebyshev polynomials', fontsize=12)
```
