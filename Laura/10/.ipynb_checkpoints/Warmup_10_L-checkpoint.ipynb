```python
# Warm-up 10

import numpy as np
import matplotlib.pyplot as plt

#Check if triplet is a pythagorean triple, evaluates equations and return boolean true/false
def pyt(a,b,c):
    return (a*a) + (b*b) == (c*c)   
```

```python
pyt(3,4,5)
```

```python
pyt(2,3,5)
```

```python
#show that if (a,b,c) is a PT, so is (qa,qb,qc)
# formally:multiply q^2 on both sides of a^2 + b^2 = c^2 assuming that (a,b,c) is a PT
# => q^2(a^2 +b^2) = q^2 c^2
# => q^2a^2 + q^2b^2 = q^2 c^2
# => (qa)^2 + (qb)^2 = (qc)^2
```

```python
# plot in (a,b) space all PT < 10000
# anything above 1000 very run time expensive (65h estimated time for 10000)

from tqdm.auto import tqdm, trange

#empty lists for collecting values of a and b that form a triplet
a_t = []
b_t = []

for a in trange(1,1000):
    for b in range (1,1000):
        for c in range (1,1000):
            if pyt(a,b,c) == True: #if the function above evaluates true, add a and b to list
                a_t.append(a)
                b_t.append(b)
plt.title('Pythagorean Triplets with a,b <1000')
plt.xlabel('a')
plt.ylabel('b')
plt.scatter(a_t,b_t,c='m',s=0.6)
```
