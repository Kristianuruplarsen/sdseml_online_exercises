```python
def TnFactory(n,x):
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        return 2*x*TnFactory(n-1,x) + TnFactory(n-2,x) 
        
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x_series = np.linspace(-1,1, 100)
# Store each series in the collumns of T_results
T_results = np.zeros([100, 5])

for i in range(5):
    T_results[:,i] = TnFactory(i+1,x_series)
    plt.plot(x_series,T_results[:,i])
plt.legend(['$T_1$','$T_2$', '$T_3$', '$T_4$', '$T_5$'])

```
