# Warm-Up: Chebyshevs Polynomials

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def TnFactory(n):
    # Base case: T0 = 1
    if n==0:
        def T(x):
            return 1
        
    # Base case: T1 = x
    elif n == 1:
        def T(x):
            return x
    
    # Recursive case:
    else:
        def T(x):
            return 2 * TnFactory(n-1) - TnFactory(n-2)
    return T(x)    
```

```python
x = 4
TnFactory(2)
```

```python
x = np.linspace(-1, 1, 100)
T1 = TnFactory(1)
T2 = TnFactory(2)
T3 = TnFactory(3)
T4 = TnFactory(4)
T5 = TnFactory(5)
```

```python
plt.figure(figsize=(12,12))
plt.plot(x, T1, label='1')
plt.plot(x, T2, label='2')
plt.plot(x, T3, label='3')
plt.plot(x, T4, label='4')
plt.plot(x, T5, label='5')
plt.legend()
```
