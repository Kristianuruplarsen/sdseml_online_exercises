>**Ex. 4.1.1:** Complete the code below to simulate data according to
$$
T = U(0,1) > 0.5 \\ 
Y(T=0) = X\beta + \epsilon \\ 
\tau(X) =  \begin{cases}
        \frac{10}{1 + e^{-\gamma X_0}} + \nu & D = 0\\ 
        \nu & D = 1
        \end{cases}\\ 
Y(T=1) = Y(0) + \tau(X) \\ 
$$
where $\epsilon, \nu$ are simply noise terms distributed according to $\mathcal{N}(0,1)$ and $\beta$ is an `N_FEATURES` vector of random parameters. $\gamma$ is a scalar parameter.


````python
N_SAMPLES = 10000
N_FEATURES = 5
GAMMA = 3
BETA = np.random.uniform(0,1, size = N_FEATURES)

X = np.random.normal(size = (N_SAMPLES, N_FEATURES)) #features
D = np.random.choice([0,1], size = N_SAMPLES) #random allocation of treatment effects

Y0 = X @ BETA + np.random.normal() # 
Tau = (1-D)*10/(1+np.exp(-GAMMA*X[:,0]))+np.random.normal(size=N_SAMPLES) #treatment effect
Y1 = Y0 + Tau
T = np.random.choice([0,1],size = N_SAMPLES) #Treatment indicator
y = Y0 + T*(Y1-Y0)
````

**Ex. 4.1.2:** Create a two-subplot figure, and plot $Y(0)$ and $Y(1)$ in one subplot against $X_0$. Plot $\tau(x)$ against $X_0$ in the other subplot. What do you see? Why do we observe $\tau=0$ in many cases?

````python
plt.figure(figsize=(20,20))
plt.subplot(221)
plt.scatter(X[:,0],Y0, label ='$Y0$')
plt.scatter(X[:,0],Y1, label ='$Y1$')
plt.xlabel('$X_0$')
plt.ylabel('$Y(T)$')
plt.legend()
plt.grid(True)

plt.subplot(222)
plt.scatter(X[:,0],Tau, label ='$Tau$')
plt.xlabel('$X_0$')
plt.ylabel('$Tau$')
plt.legend()
plt.grid(True)

plt.show()
````

$\tau=0$ in many cases because D and $\nu$ is drawn from a standard normal distribution. This means that there is no treatment effect.
