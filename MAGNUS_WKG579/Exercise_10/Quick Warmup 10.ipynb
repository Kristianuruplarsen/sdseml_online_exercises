If *(a, b, c)* is a Pythagorean triple, then so is *(qa, qb, qc)* because it can be reduced to the same form as *(a, b, c)* using only multiplication and division:

\begin{align}
(qa)^2 + (qb)^2 & = (qc)^2 \\
q^2a^2 + q^2b^2 & = q^2c^2 \\
a^2 + b^2 & = c^2 \\
\end{align}

```python
a_list, b_list = [], []
for a in range(1, 10000):
    for b in range(1, 10000):
        if (((a)*(a) + (b)*(b))**0.5)%1 == 0:
            a_list.append(a)
            b_list.append(b)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(12,8))

ax.scatter(a_list, b_list, color='r',s=0.01, alpha=0.5)

ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_title('Scatter plot of all Pythagorean triples')
plt.show()
```
