```python
from numpy import arange
import matplotlib.pyplot as plt

# define recursive function

def TnFactory(n, x):
    
    # special cases for low n
    if n == 0:
        return 1
    
    elif n == 1:
        return x
    
    # general case
    else:
        return 2*x*TnFactory(n-1,x)-TnFactory(n-2,x)

# for each n 0 - 5, calculate list of T_n(order)

num_range = arange(-1,1.01,0.01)

poly_dict = {}

n_max = 5

for n in range(0,n_max+1):
    
    n_list = []
    
    for i in num_range:
        
        n_list.append(TnFactory(n, i))
        
    poly_dict[n] = n_list
    
    
# create a figure

fig = plt.figure(figsize=(16,10))
ax = plt.axes()


for n in range(0,n_max+1):
    ax.plot(num_range, poly_dict[n], label = f'n = {n}')

plt.xlabel('x', fontsize=30) # large text this time
plt.xticks(fontsize=15) # large text this time
plt.ylabel('$T_n(x)$', fontsize=30) # large text this time
plt.yticks(fontsize=15) # large text this time
plt.legend()
plt.show()
```
