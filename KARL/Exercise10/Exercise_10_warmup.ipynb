```python
import math
from matplotlib import pyplot as plt

i = 0
a_ls = []
b_ls = []
for a in range(1,10000):
    for b in range(1,10000):
        if math.sqrt(a**2 + b**2).is_integer():
            i += 1
            a_ls.append(a)
            b_ls.append(b)

print('{} Pythagorean triples were found'.format(i))
plt.scatter(a_ls, b_ls, marker='.', s=0.5)
plt.xlabel('a')
plt.ylabel('b')
```

To show that (qa, qb, qc) is a Pythagorean triple, if (a, b, c) is, we start from (a, b, c), mulitply by $q^2$ on both sides and reduce:
$ a^2 + b^2 = c^2 \Leftrightarrow q^2\cdot a^2 + q^2 \cdot b^2 = q^2 \cdot c^2 \Leftrightarrow (qa)^2 + (qb)^2 = (qc)^2 $
