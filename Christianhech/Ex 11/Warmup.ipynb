```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
plt.style.use('ggplot')

def TnFacctory(n):
    """This is a recursive function to replicate Chebychevs polynomials"""
    if n == 0:
        def T_n(x):
            return 1
    elif n ==1:
        def T_n(x):
            return x
    else:
        def T_n(x):
            return 2*x*TnFacctory(n-1)(x)-TnFacctory(n-2)(x)
    return T_n


N = 5 # number of polynomials
x = np.linspace(-1,1,1000)
plt.figure(figsize=(6,4),dpi=600)
for i in range(1,N+1):
    plt.plot(x, TnFactory(i)(x), label='$T_{}(x)$'.format(i))
plt.xlabel('x')
plt.ylabel('$T_n(x)$')
plt.legend(loc='best')
plt.title('The first 5 Chebyshev polynomials$', fontsize=12)
plt.show()
```
