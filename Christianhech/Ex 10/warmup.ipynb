```python
import math  

#Q1:

'''
It is given that if (a,b,c) is a Pythagorean triple, so is
(ka,kb,kc) for any positive integer K:

(Ka)^2 + (Kb)^2 = K^2 * a^2 + K^2 * b^2 = K^2(a^2+b^2)

If we set this equal to (Kc)^2 = K^2 * c^2, then we find:

K^2(a^2+b^2) = K^2 * c^2 => a^2+b^2 = c^2

'''

#We can see/test this easily:


def pythagorean(a,b,k):
    c = math.sqrt(a**2 + b**2)
    c2 = (k*a)**2 + (k*b)**2
    if round((k*c)**2,10) == c2:
        return True
# Test
print('Is (ka,kb,kc) for a positive integer k an pythagorean triple? :', pythagorean(2,3,10))

#Q2 - Plotting the (a,b) space

# We define a function that determines of (a,b,c) is a pythagorean triple:
def check(a,b,c):
    if a**2+b**2 == c**2:
        return True

a_list = []
b_list = []
for a_i in range(500):
    for b_i in range(500):
        for c_i in range(500):
            if check(a_i,b_i,c_i):
                a_list.append(a_i)
                b_list.append(b_i)


import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use('ggplot')
plt.scatter(a_list,b_list)
plt.xlabel('a')
plt.ylabel('b')
plt.title('(a,b) space for all Pythagorean triples')
plt.show()
```
