To show that it is a Pythagorean triple, we multiply $q^2$ onto both sides of the equality. Thus we get:
$$a^2 + b^2 = c^2$$
$$q^2(a^2 + b^2) = q^2c^2$$
$$q^2a^2 + q^2b^2 = q^2c^2$$
$$(qa)^2 + (qb)^2 = (qc)^2$$
Which exactly is a Pythagorean triple as well.

```python
import math
from matplotlib import pyplot as plt

i_ = 0
a_t = []
b_t = []
for a in range(1,10000):
    for b in range(1,10000):
        if math.sqrt(a**2 + b**2).is_integer():
            i_ += 1
            a_t.append(a)
            b_t.append(b)

print(i_, 'Pythagorean triples were found for $a,b<10.000$')
plt.scatter(a_t, b_t, color = 'green', marker = '.', s = 0.4)
plt.xlabel('a')
plt.ylabel('b')
```
